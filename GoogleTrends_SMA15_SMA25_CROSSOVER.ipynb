```python
import matplotlib.pyplot as plt 
import pandas as pd 
import yfinance as yf 
import numpy as np
from pytrends.request import TrendReq
pytrends = TrendReq()
```

```python
amc = yf.download('WISH', '2020-03-01', '2021-06-22')
amc = amc['Adj Close']
amc
```

```python
kw_list = ['WISH']

pytrends.build_payload(kw_list, timeframe='today 12-m')
pytrends
```

```python
data = pytrends.interest_over_time()
data
```

```python
SMA15 = pd.DataFrame()
SMA15['SMA 15'] = data['WISH'].rolling(window= 15).mean()
SMA15
```

```python
SMA25 = pd.DataFrame()
SMA25['SMA 25'] = data['WISH'].rolling(window= 25).mean()
SMA25
```

```python
plt.figure(figsize=(12.5, 4.5))
plt.plot(data['WISH'], label = 'WISH')
plt.plot(SMA15['SMA 15'], label = 'SMA15')
plt.plot(SMA25['SMA 25'], label = 'SMA25')
plt.title('WISH Google Trends w/ SMA15 SMA25')
plt.xlabel('2020-06-01 - 2021-06-22')
plt.ylabel('Interest Overtime')
plt.legend(loc='upper left')
```

```python
data2 = pd.DataFrame()
data2['WISH'] = data['WISH']
data2['SMA15']= SMA15['SMA 15']
data2['SMA25'] = SMA25['SMA 25']
data2
```

```python
def buy_sell(data):
    sigPriceBuy = []
    sigPriceSell= []
    flag = -1
    
    for i in range(len(data)):
        if data['SMA15'][i] > data['SMA25'][i]:
            if flag != 1:
                sigPriceBuy.append(data['WISH'][i])
                sigPriceSell.append(np.nan)
                flag = 1
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        elif data['SMA15'][i] < data['SMA25'][i]:
            if flag != 0:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(data['WISH'][i])
                flag = 0
            else:
                sigPriceBuy.append(np.nan)
                sigPriceSell.append(np.nan)
        else: 
            sigPriceBuy.append(np.nan)
            sigPriceSell.append(np.nan)
    return (sigPriceBuy, sigPriceSell)
```

```python
buy_sell = buy_sell(data2)
data2['Buy_Signal_Price'] = buy_sell[0]
data2['Sell_Signal_Price'] = buy_sell[1]
```

```python
data2.head(1)
```

```python
plt.figure(figsize=(12.6, 4.6))
plt.plot(data2['WISH'], label = 'WISH', alpha = 0.35)
plt.plot(data2['SMA15'], label = 'SMA15', alpha = 0.35)
plt.plot(data2['SMA25'], label = 'SMA25', alpha = 0.35)
plt.scatter(data2.index, data2['Buy_Signal_Price'], label = 'Buy', marker = '^', color = 'green')
plt.scatter(data2.index, data2['Sell_Signal_Price'], label = 'Sell', marker='v', color = 'red')
plt.title('WISH Google Trends SMA15 SMA25 Crossovers')
plt.xlabel('2020-06-01 - 2021-06-22')
plt.ylabel('Interest Overtime')
plt.legend(loc='upper left')
plt.show()
```
